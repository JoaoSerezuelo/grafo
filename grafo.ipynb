# Grafo

- NetworkX: manipulação de grafo
- Matplotlib: visualização

```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Exemplos de Grafos:

1. **Grafo Não Orientado**

2. **Grafo Não Orientado com Peso**``

3. **Grafo Orientado**

4. **Grafo Orientado com Peso**


```python
# Criar um grafo não orientado
G = nx.Graph()

# Adicionar nós
G.add_nodes_from([1, 2, 3, 4])

# Adicionar arestas
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])

# Visualizar o grafo
pos = nx.spring_layout(G)  # Layout do grafo
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', width=2)
plt.title("Grafo Não Orientado Simples")
plt.show()
```

```python
# Grafo não orientado com pesos
G3 = nx.Graph()

# Adicionando nós
G3.add_nodes_from([1, 2, 3, 4])

# Adicionando arestas com pesos
G3.add_weighted_edges_from([(1, 2, 4), (1, 3, 2), (2, 3, 3), (3, 4, 1)])

# Desenhando o grafo
pos = nx.spring_layout(G3)
edge_labels = {(u, v): d['weight'] for u, v, d in G3.edges(data=True)}
nx.draw(G3, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', width=2)
nx.draw_networkx_edge_labels(G3, pos, edge_labels=edge_labels)
plt.title("Grafo Não Orientado com Pesos")
plt.show()
```

```python
# Grafo orientado simples
G2 = nx.DiGraph()

# Adicionando nós
G2.add_nodes_from([1, 2, 3, 4])

# Adicionando arestas
G2.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Desenhando o grafo
pos = nx.spring_layout(G2)
nx.draw(G2, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', width=2, arrows=True)
plt.title("Grafo Orientado Simples")
plt.show()
```

```python
# Grafo orientado com pesos
G4 = nx.DiGraph()

# Adicionando nós
G4.add_nodes_from([1, 2, 3, 4])

# Adicionando arestas com pesos
G4.add_weighted_edges_from([(1, 2, 4), (1, 3, 2), (2, 3, 3), (3, 4, 1)])

# Desenhando o grafo
pos = nx.spring_layout(G4)
edge_labels = {(u, v): d['weight'] for u, v, d in G4.edges(data=True)}
nx.draw(G4, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', width=2, arrows=True)
nx.draw_networkx_edge_labels(G4, pos, edge_labels=edge_labels)
plt.title("Grafo Orientado com Pesos")
plt.show()
```

## DISTANCIA

## Excentricidade de um vértice

```python
#não orientado

# Calculando excentricidade de um vértice
#no lugar do G colocar o grafo que deseja e do v o vértice que deseja
excentricidade = nx.eccentricity(G3, v=2)
print("Excentricidade do vértice:", excentricidade)
```

```python
#orientado

# Calculando excentricidade de um vértice
#no lugar do G colocar o grafo que deseja e do v o vértice que deseja
excentricidade = nx.eccentricity(G2.reverse(), v=3)
print("Excentricidade do vértice 3:", excentricidade)
```

## não orientado e orientado -- Diametro, raio e centro 

```python
#Diametro
diametro = nx.diameter(G3)
print("Diâmetro do grafo:", diametro)
# Raio do grafo
raio = nx.radius(G3)
print("Raio do grafo:", raio)
# Centro do grafo
centro = nx.center(G3)
print("Centro do grafo:", centro)
```

## vertices perifericos

```python
#não orientado
excentricidades = nx.eccentricity(G3)
perifericos = [v for v, excentricidade in excentricidades.items() if excentricidade == diametro]
print("Vértices periféricos:", perifericos)
```

```python
#orientado
excentricidades = nx.eccentricity(G3.reverse())
perifericos = [v for v, excentricidade in excentricidades.items() if excentricidade == diametro]
print("Vértices periféricos:", perifericos)
```

## exemplo

```python
# Criar um grafo não orientado
G = nx.Graph()

# Adicionar nós
G.add_nodes_from([1, 2, 3, 4, 5, 6])

# Adicionar arestas
G.add_edges_from([(1, 2), (1, 3), (1,4), (4,1), (2,4), (2,3), (2,5), (4,5), (3,5), (3,6), (6,5)])

# Visualizar o grafo
pos = nx.spring_layout(G)  # Layout do grafo
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_color='black', width=2)
plt.title("Grafo Não Orientado Simples")
plt.show()
```

```python
# Calculando a matriz de distâncias entre todos os pares de vértices
distancias = dict(nx.all_pairs_shortest_path_length(G))

# Calculando a excentricidade de todos os vértices
excentricidades = nx.eccentricity(G)

# Imprimindo a matriz de distâncias e excentricidade
print("Matriz de Distâncias e Excentricidade:")
for origem, distancias_destino in distancias.items():
    linha = ""
    for destino in range(1, 7):  # Nós de 1 a 6
        if destino in distancias_destino:
            distancia = distancias_destino[destino]
            linha += f"{distancia} "
        else:
            linha += "Inf "  # Se não há caminho, coloca Infinito
    linha += f"  Excentricidade: {excentricidades[origem]}"
    print(linha)
```
